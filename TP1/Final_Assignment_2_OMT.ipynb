Create a list showing the total number of medals won for each country over the history of the Olympics. For each country, include the year of the first and most recent Olympic medal wins. (3 points)

```python
medal_counts = df_olympic_wna.groupby('NOC').agg(
    Total_Medals=pd.NamedAgg(column='Medal', aggfunc='count'),
    First_Year=pd.NamedAgg(column='Edition', aggfunc='min'),
    Most_Recent_Year=pd.NamedAgg(column='Edition', aggfunc='max')
).reset_index()

print(medal_counts)
```
