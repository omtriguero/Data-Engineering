2)	Read the daily confirmed cases and deaths into two dataframes

```python
import pandas as pd
import numpy as np
```

```python
#reading the data
confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

df_confirmedcases = pd.read_csv(confirmed_cases_url)
df_deaths = pd.read_csv(deaths_url)
```

```python
pd.DataFrame(df_confirmedcases)
```

```python
pd.DataFrame(df_deaths)
```

3)	Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

```python
# Deleting unneccessary columns

del df_confirmedcases['Province/State']
del df_confirmedcases['Country/Region']
del df_confirmedcases['Lat']
del df_confirmedcases['Long']
```

```python
# Deleting unneccessary columns

del df_deaths['Province/State']
del df_deaths['Country/Region']
del df_deaths['Lat']
del df_deaths['Long']
```

```python

# Aggregate and sums confirmed cases  on a global level

confirmed_cases = df_confirmedcases.sum(axis=0)
confirmed_cases.index.name = 'Date'
confirmed_cases.name = 'confirmed_cases'

confirmed_cases_final = pd.DataFrame(confirmed_cases)


confirmed_cases_final.reset_index()

confirmed_cases_final
```

```python
# Aggregate and sum deaths on a global level

covid_death = df_deaths.sum(axis=0)
covid_death.index.name = 'Date'

covid_death.name = 'deaths'
covid_death_final = pd.DataFrame(covid_death)
covid_death_final
```

```python
# Join both the cases and deaths into a single dataframe
final_covid_df = confirmed_cases_final.join(covid_death_final)
```

```python
# Save the dates as a column and set it as an index

final_covid_df["Date"]= final_covid_df.index
final_covid_df.set_index("Date", inplace=True, drop=True)
```

```python
pd.DataFrame(final_covid_df)
```
